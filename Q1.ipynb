1 - Analisar a imagem de entrada com o objetivo de aplicar algum pré-processamento para redução de ruído. Utilize livremente as funções existentes na OpenCV.

Para redução dos ruídos das imagens foi utilizada o filtro da mediana. A dimensão da máscara utilizada é de 5x5


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("test_images/bolhas.png") #leitura da imagem de entrada
```

```python
median = cv2.medianBlur(img,5) # aplicação do filtro de mediana com mascara 5x5
```

```python
#plotagem das imagens geradas
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title('Mediana')
plt.xticks([]), plt.yticks([])
plt.show()
```
